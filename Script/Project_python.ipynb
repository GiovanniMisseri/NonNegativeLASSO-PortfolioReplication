```python
import pandas as pd
import numpy as np
```

```python
#dataset (data trasformed in daily log-returns)
dataset=pd.read_csv("./dldat.csv",index_col=0)
dataset
```

```python
mask=~dataset.isnull().any()
```

```python
#delete varaibles with nan values
dataset=dataset.loc[:,mask]
dataset
```

```python
#divide dataset into X and y
y=dataset.iloc[:,0].values
X=dataset.iloc[:,1:].values
```

```python
import sklearn.linear_model 
```

```python
#replication errors varying the regularization parameter (~ varying the number of variables)
window=250
columns=['lambda','#coeff','abs_err', 'rel_error']
alpha_set=np.linspace(0.0001,0.01,10)
alpha_table=pd.DataFrame(data=np.zeros((len(alpha_set),len(columns))),columns=columns)
for i,alpha in enumerate (alpha_set):
    mod=sklearn.linear_model.Lasso(alpha=alpha, positive=True)
    mod.fit(sqrt(250)*X[:window,:],sqrt(250)*y[:window])
    pred=mod.predict(X[:window,:])
    abs_err=np.mean(abs(y[:window]-pred))
    abs_err100=abs(y[window:window+100]-mod.predict(X[window:window+100,:]))
    abs_err50=np.mean(abs(y[window:window+100]-mod.predict(X[window:window+100,:])))
    
    rel_err=np.zeros(len(y[:window]))
    rel_err50=np.zeros(len(y[window:window+100]))
    
    for j in range(len(y[:window])):
        if y[j]!=0:
            rel_err[j]=abs(y[:window]-pred)[j]/abs(y[j])
        else:
            rel_err[j]=np.NaN
    
    for j in range(len(y[window:window+100])):
        if y[window+j]!=0:
            rel_err50[j]=abs_err100[j]/abs(y[window+j])
        else:
            rel_err50[j]=np.NaN
    alpha_table.loc[i,'lambda']=alpha
    alpha_table.loc[i,'abs_err']=abs_err
    alpha_table.loc[i,'rel_error']=np.nanmean(rel_err)
    alpha_table.loc[i,'#coeff']=np.count_nonzero(mod.coef_)
alpha_table
```

```python
#beta coefficients
window=250
mod_beta=sklearn.linear_model.Lasso(alpha=0.000003, positive=True)
beta_dataset=pd.DataFrame()
for k in range(558):
    mod_beta.fit(X[k:k+window,:],y[k:k+window])
    beta_dataset[k]=mod_beta.coef_
beta_dataset=np.transpose(beta_dataset)

coeff_count=np.zeros(beta_dataset.shape[0])
for i in range(beta_dataset.shape[0]):
    coeff_count[i]=np.count_nonzero(beta_dataset.iloc[i,:])
beta_dataset['number_coeff']=coeff_count
beta_dataset.insert(0,'intercept',mod_beta.intercept_)
beta_dataset
```

```python
#passive approach varying number of variables
start=250
end=800
window=250
alpha_set=[0.00000002,0.0000002,0.0000008,0.000002,0.000003,0.000005,0.000008,0.00001,0.00002]
budget=1
fixed_fee=0.00002*budget
perc_fee=0.002*budget
portfolio_pass_alpha=pd.DataFrame(data=np.zeros((len(alpha_set),3)),columns=['lambda','#coeff','portfolio_value'])
for i,alpha in enumerate(alpha_set):
    mod_pass_alpha=sklearn.linear_model.Lasso(alpha=alpha, positive=True)
    mod_pass_alpha.fit(X[start-window:start,:],y[start-window:start])
    budget=1*(1-perc_fee)-fixed_fee*np.count_nonzero(mod_pass_alpha.coef_)
    portfolio_gain=mod_pass_alpha.predict(X[start:end,:])
    portfolio_value=budget*np.exp(sum(portfolio_gain))
    portfolio_pass_alpha.loc[i,'lambda']=alpha
    portfolio_pass_alpha.loc[i,'#coeff']=np.count_nonzero(mod_pass_alpha.coef_)
    portfolio_pass_alpha.loc[i,'portfolio_value']=portfolio_value
portfolio_pass_alpha
```

```python
#passive approach with 69 variables
start=250
end=800
window=250
alpha=8.000000e-06
budget=1
fixed_fee=0.00002*budget
perc_fee=0.002*budget
portfolio_pass=pd.DataFrame(data=np.zeros((end-start+1,4)),columns=['y','portfolio_gain','value_index','value_portfolio'],index=range(249,800,1))
mod_pass=sklearn.linear_model.Lasso(alpha=alpha, positive=True)
mod_pass.fit(X[start-window:start,:],y[start-window:start])
budget=budget*(1-perc_fee)-fixed_fee*np.count_nonzero(mod_pass.coef_)
portfolio_pass.loc[249,'y']=np.nan
portfolio_pass.loc[249,'portfolio_gain']=np.nan
portfolio_pass.loc[249,'value_index']=1
portfolio_pass.loc[249,'value_portfolio']=budget
portfolio_pass.loc[250:800,'y']=y[250:800]
portfolio_pass.loc[250:800,'portfolio_gain']=mod_pass.predict(X[start:end,:])
for i in range(250,800,1):
    portfolio_pass.loc[i,'value_index']=portfolio_pass.loc[i-1,'value_index']*np.exp(portfolio_pass.loc[i,'y'])
    portfolio_pass.loc[i,'value_portfolio']=portfolio_pass.loc[i-1,'value_portfolio']*np.exp(portfolio_pass.loc[i,'portfolio_gain'])
portfolio_pass
```

```python
#active approach with logistic regression
window=250
window_pred=100
alpha=8.000000e-06
delta=0.002
start=250
portfolio=pd.DataFrame(data=np.zeros((0,4)),columns=['y','y_0/1','pred_y','real_gain'])
y_disc=y.copy()+np.log(1-delta)
y_logistic=y_disc.copy()
y_logistic[y_logistic>0]=1
y_logistic[y_logistic<=0]=0
for t in range(start,len(dataset),1):
    mod=sklearn.linear_model.Lasso(alpha=alpha, positive=True)
    mod.fit(X[t-window:t,:],y_disc[t-window:t])
    mod_pred=sklearn.linear_model.LogisticRegression(penalty='l2',C=100)
    mod_pred.fit(X[t-window_pred-1:t-1,:],y_logistic[t-window_pred:t])
    portfolio.loc[t,'y']=y_disc[t]
    portfolio.loc[t,'y_0/1']=y_logistic[t]
    portfolio.loc[t,'pred_y']=mod_pred.predict(X[t-1:t,:])
    portfolio.loc[t,'real_gain']=mod.predict(X[t:t+1,:])
print(portfolio)


portfolio_values=[budget]
for i,period in enumerate(range(start,len(dataset),1)):
    if portfolio.pred_y[period]==1:
        portfolio_values.append(portfolio_values[i]*np.exp(portfolio.real_gain[period]))
    else: portfolio_values.append(portfolio_values[i])
portfolio_values
```

```python
#predictions statistics
true_pos=0
true_neg=0
false_pos=0
false_neg=0
for i in range(250,808,1):
    if (portfolio.loc[i,'y_0/1']==1) and (portfolio.loc[i,'pred_y']==1):
        true_pos+=1
for i in range(250,808,1):
    if (portfolio.loc[i,'y_0/1']==0) and (portfolio.loc[i,'pred_y']==0):
        true_neg+=1
for i in range(250,808,1):
    if (portfolio.loc[i,'y_0/1']==0) and (portfolio.loc[i,'pred_y']==1):
        false_pos+=1
for i in range(250,808,1):
    if (portfolio.loc[i,'y_0/1']==1) and (portfolio.loc[i,'pred_y']==0):
        false_neg+=1
```

```python
#active approach with LASSO predictions
window=250
window_pred=100
alpha=8.000000e-06
delta=0.002
start=250
portfolio2=pd.DataFrame(data=np.zeros((0,3)),columns=['y','pred_y','real_gain'])
y_disc=y.copy()+np.log(1-delta)
for t in range(start,len(dataset),1):
    mod2=sklearn.linear_model.Lasso(alpha=alpha, positive=True)
    mod2.fit(X[t-window:t,:],y[t-window:t])
    mod_pred2=sklearn.linear_model.Lasso(alpha=alpha, positive=True)
    mod_pred2.fit(X[t-window_pred-1:t-1,:],y_disc[t-window_pred:t])
    portfolio2.loc[t,'y']=y_disc[t]
    portfolio2.loc[t,'pred_y']=mod_pred2.predict(X[t-1:t,:])
    portfolio2.loc[t,'real_gain']=mod2.predict(X[t:t+1,:])
print(portfolio2)


portfolio_values2=[budget]
for i,period in enumerate(range(start,len(dataset),1)):
    if portfolio2.pred_y[period]>0:
        portfolio_values2.append(portfolio_values2[i]*np.exp(portfolio2.real_gain[period]))
    else: portfolio_values2.append(portfolio_values2[i])
portfolio_values2
```
