```python
window=250
alpha=0.000015
tab=pd.DataFrame(data=np.zeros((window,0)))
mod=sklearn.linear_model.Lasso(alpha=alpha, positive=True)
mod.fit(X[:window,:],y[:window])
pred=mod.predict(X[:window,:])
abs_error=abs(y[:window]-pred)
tab['y']=y[:window]
tab['predictions']=pred
tab['abs_err']=abs_error
#tab['rel_error']=np.divide(abs_error,y[:window])
tab['coeff']=np.count_nonzero(mod.coef_)
tab
```

```python
window=250
columns=['alpha','coeff','abs_err', 'rel_error','abs_err[t,t+50]','abs_err[t,t+50]','rel_err[t,t+50]']
alpha_set=np.linspace(0.0000001,0.0001,20)
alpha_table=pd.DataFrame(data=np.zeros((len(alpha_set),len(columns))),columns=columns)
for i,alpha in enumerate (alpha_set):
    mod=sklearn.linear_model.Lasso(alpha=alpha, positive=True)
    mod.fit(X[:window,:],y[:window])
    pred=mod.predict(X[:window,:])
    abs_error=np.mean(abs(y[:window]-pred))
    abs_error20=np.mean(abs(y[window:window+50]-mod.predict(X[window:window+50,:])))
    alpha_table.loc[i,'alpha']=alpha
    alpha_table.loc[i,'abs_err']=abs_error
    alpha_table.loc[i,'rel_error']=abs_error/np.mean(y[:window])
    alpha_table.loc[i,'coeff']=np.count_nonzero(mod.coef_)
    alpha_table.loc[i,'abs_err[t,t+50]']=abs_error20
    alpha_table.loc[i,'rel_err[t,t+50]']=abs_error20/np.mean(y[window:window+50])
alpha_table
```

```python
window=500
columns=['alpha','coeff','abs_err', 'rel_error','abs_err[t,t+100]','rel_err[t,t+100]']
alpha_set=np.linspace(0.00001,0.0001,20)
alpha_table=pd.DataFrame(data=np.zeros((len(alpha_set),len(columns))),columns=columns)
for i,alpha in enumerate (alpha_set):
    mod=sklearn.linear_model.Lasso(alpha=alpha, positive=True)
    mod.fit(X.iloc[:window,:],y[:window])
    pred=mod.predict(X.iloc[:window,:])
    abs_err=np.mean(abs(y[:window]-pred))
    abs_err50=np.mean(abs(y[window:window+100]-mod.predict(X.iloc[window:window+100,:])))
    
    rel_err=np.zeros(len(y[:window]))
    rel_err50=np.zeros(len(y[window:window+100]))
    
    for j in range(len(y[:window])):
        if y.iloc[j]!=0:
            rel_err[j]=abs(y[:window]-pred).iloc[j]/abs(y.iloc[j])
        else:
            rel_err[j]=np.NaN
    
    for j in range(len(y[window:window+100])):
        if y.iloc[window+j]!=0:
            rel_err50[j]=abs(y[window:window+100]-mod.predict(X.iloc[window:window+100,:])).iloc[j]/abs(y.iloc[window+j])
        else:
            rel_err50[j]=np.NaN
    alpha_table.loc[i,'alpha']=alpha
    alpha_table.loc[i,'abs_err']=abs_err
    alpha_table.loc[i,'rel_error']=np.nanmean(rel_err)
    alpha_table.loc[i,'coeff']=np.count_nonzero(mod.coef_)
    alpha_table.loc[i,'abs_err[t,t+100]']=abs_err50
    alpha_table.loc[i,'rel_err[t,t+100]']=np.nanmean(rel_err50)
alpha_table
```

```python
window=250
mod_beta=sklearn.linear_model.Lasso(alpha=0.000015, positive=True)
beta_dataset=pd.DataFrame()
for k in range(558):
    mod_beta.fit(X.iloc[k:k+window,:],y[k:k+window])
    beta_dataset[k]=mod_beta.coef_
beta_dataset=np.transpose(beta_dataset)

coeff_count=np.zeros(beta_dataset.shape[0])
for i in range(beta_dataset.shape[0]):
    coeff_count[i]=np.count_nonzero(beta_dataset.iloc[i,:])
beta_dataset['number_coeff']=coeff_count
beta_dataset.insert(0,'intercept',mod_beta.intercept_)
beta_dataset
```

```python
window=250
step=1
period_set=range(start,stop,step)
alpha=0.00001
portfolio=pd.DataFrame(data=np.zeros((window,1)),columns=['pred'])
for t in range(window,len(dataset)):
    if t in period_set:
        mod=sklearn.linear_model.Lasso(alpha=alpha, positive=True)
        mod.fit(X[t-window:t,:],y[t-window:t])
    portfolio.loc[t,'pred']=mod.predict(X[t:t+1,:])


delta=0.002
budget=1
portfolio_values=[budget]
for i,period in enumerate(period_set):
    if period+step<dataset.shape[0]:
        if sum(portfolio.pred[period:period+step])>0:
            portfolio_values.append(portfolio_values[i]*(1-delta)*np.exp(sum(portfolio.pred[period:period+step])))
        else: portfolio_values.append(portfolio_values[i])
    else:
        if sum(portfolio.pred[period:])>0:
            portfolio_values.append(portfolio_values[i]*(1-delta)*np.exp(sum(portfolio.pred[period:])))
        else: portfolio_values.append(portfolio_values[i])
portfolio_values
```
